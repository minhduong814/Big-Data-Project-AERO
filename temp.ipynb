```python
import requests

API_KEY = "YOUR_API_KEY_HERE"
BASE_URL = "http://api.aviationstack.com/v1/flights"

params = {
    "access_key": API_KEY,
    "limit": 10  # Get 10 flights
}

response = requests.get(BASE_URL, params=params)

if response.status_code == 200:
    data = response.json()
    print(f"✅ API Working! Found {data['pagination']['total']} flights")
    print(f"Monthly limit: {data['pagination']['limit']}")
else:
    print(f"❌ Error: {response.status_code}")
    print(response.text)
```
